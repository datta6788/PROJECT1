```python
import joblib
import pandas as pd 
from scipy.sparse import load_npz
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sparse_matrix=load_npz(r"D:\PRJ1\PROJECT1\models\sparse_matrix.npz")
```

```python
ani_to_index=joblib.load(r"D:\PRJ1\PROJECT1\models\ani_to_index.pkl")
```

```python
tsvd=TruncatedSVD(n_components=50,random_state=42)
anime_emb=tsvd.fit_transform(sparse_matrix)
```

```python
tsvd_sim=cosine_similarity(anime_emb)
```

```python
anime_emb.shape
```

```python
ani_names=list(ani_to_index.keys())
index_tsvd=pd.Series(range(len(ani_names)),index=ani_names)
```

```python
def tsvd(anime_name,no_recomm=10):
    index=index_tsvd[anime_name]
    similarity_scores=list(enumerate(tsvd_sim[index]))
    similarity_scores=sorted(similarity_scores,key=lambda x:x[1],reverse=True)
    similarity_scores=similarity_scores[1:no_recomm+1]
    recomm=[]
    for ani_idx,score in similarity_scores:
        recomm.append({"Name": ani_names[ani_idx],"Score": float(score)})
    return pd.DataFrame(recomm)
```

```python
tsvd("death note",no_recomm=20)
```
